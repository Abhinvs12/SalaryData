# Sorting And Comparison Operations

```python
# Calculating the average salary in each year

df.groupby("work_year")["salary"].mean()
```

```python
# To get the average and maximum salary for each job in every year

df.groupby(["work_year","job_title"]).agg(
avg_salary = ("salary", "mean"),
max_salary = ("salary", "max")
)
```

```python
# To get the no of groups

df.groupby(["work_year","job_title"]).ngroups
```

```python
# To retrieve data of a particular case
# To get the details of Data Engineers working in US

data_engineer_us=df.groupby(["job_title","company_location"]).get_group(("Data Engineer","US"))
data_engineer_us.head()
```

```python
# Retrieving the details of those having both the company location and residence as US

df.loc[(df["company_location"]=="US") & (df["employee_residence"]=="US")]
```

```python
# Retrieving the details of those having salary>400000 and salary currency equals GBP

df.loc[(df["salary_in_usd"] >= 400000) & (df["salary_currency"]=="GBP")]
```

```python
# Retrieving the details of consultant data engineer and data quality engineer

df.loc[(df["job_title"]=="Consultant Data Engineer") | (df["job_title"]=="Data Quality Engineer")]
```

```python
# Retrieving the details of those having experience level as SE and employment type CT

df.loc[(df["experience_level"]=="SE") & (df["employment_type"]=="CT")]
```

# Graphical Representations

```python
# Plotting a boxplot to detect outliers in salary

sns.boxplot(x=df["salary_in_usd"])
```

```python
# Plotting a bar graph that gives the count of employees in each field

df.job_title.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5),colormap="winter_r")
plt.title("Number of employees  in each field")
plt.ylabel("No of employees")
plt.xlabel("Jobs")
```

```python
# Plotting heatmap to know how the variables are correlated

plt.figure(figsize=(10,5))
h=df.corr()
sns.heatmap(h,cmap="BrBG",annot=True)
h
```

```python
# Plotting a piechart for company_size

p=df["company_size"].value_counts()
print(p)
mylabels = ["Medium","Large","Small"]
myexplode=[0.1,0.04,0]
plt.pie(p,labels=mylabels,shadow=True,explode=myexplode,autopct="%.0f%%")
plt.legend(title="Company size")
plt.show()
```

```python
# Subplot

# first subplot

plt.subplot(1,2,1)
sns.barplot(x="employment_type",y="salary_in_usd",data=df)
plt.title("Subplot 1")
plt.xlabel("Employment_type")
plt.ylabel("Salary")

#second subplot

plt.subplot(1,2,2)
sns.barplot(x="experience_level",y="salary_in_usd",data=df)
plt.title("Subplot 2")
plt.xlabel("Experience level")
plt.ylabel("Salary")
plt.show()
```

```python
# Plotting barplot to get the average salary in each year

sns.lineplot(x="work_year",y="salary_in_usd",data=df)
plt.title("Average salary in each year")
plt.xlabel("Year")
plt.ylabel("Salary")
plt.show()
```

```python
# Plotting the histogram for various fields 

new_df=df[["experience_level","employment_type","salary_currency","company_size"]]
# print(new_df)
sns.set()
fig,axes=plt.subplots(2,2)
sns.histplot(new_df,x="experience_level",ax=axes[0,0])
sns.histplot(new_df,x="employment_type",ax=axes[0,1])
sns.histplot(new_df,x="salary_currency",ax=axes[1,0])
sns.histplot(new_df,x="company_size",ax=axes[1,1])
plt.show()
    
```

```python

# Plotting a bargraph that gives the average salary for each job

salary_df=df[["job_title","salary_in_usd"]].head(50)
sdf=pd.DataFrame(salary_df)
sns.barplot(y="job_title",x="salary_in_usd",data=sdf)
plt.title("Salary for each job")
plt.xlabel("Salary")
plt.ylabel("Job Type")

```
